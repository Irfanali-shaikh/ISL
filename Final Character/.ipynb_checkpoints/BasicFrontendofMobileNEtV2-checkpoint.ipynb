```python
MODEL_PATH = r'E:\M.Tech\Dissertation Stage 2 - Final\Final Character\ISL_mobileNET.h5'

# Load your trained model
model = load_model(MODEL_PATH)
```

```python
def model_predict(img_path, model):
    print(img_path)
    img = image.load_img(img_path, target_size=(224, 224))
    # Preprocessing the image
    x = image.img_to_array(img)
    # x = np.true_divide(x, 255)
    ## Scaling
    x=x/255
    x = np.expand_dims(x, axis=0)
   

    

    preds = model.predict(x)
    print(preds)
    preds=np.argmax(preds, axis=1)
    if preds==0:
        preds="A"
    elif preds==1:
        preds="B"
    elif preds==2:
        preds="C"
    elif preds==3:
        preds="D"
    elif preds==4:
        preds="E"
    elif preds==5:
        preds="F"
    elif preds==6:
        preds="G"
    elif preds==7:
        preds="H"
    elif preds==8:
        preds="I"
    elif preds==9:
        preds = "J"
    elif preds==10:
        preds = "K"
    elif preds==11:
        preds = "L"
    elif preds==12:
        preds = "M"
    elif preds==13:
        preds = "N"
    elif preds==14:
        preds = "O"
    elif preds==15:
        preds = "P"
    elif preds==16:
        preds = "Q"
    elif preds==17:
        preds = "R"
    elif preds==18:
        preds = "S"
    elif preds==19:
        preds = "Space"
    elif preds==20:
        preds = "T"
    elif preds==21:
        preds = "U"
    elif preds==22:
        preds = "V"
    elif preds==23:
        preds = "W"
    elif preds==24:
        preds = "X"
    elif preds==25:
        preds = "Y"
    elif preds==26:
        preds = "Z"
    else:
        preds="Not a Character"
        
    
    
    return preds
```

```python
file_path = r"E:\M.Tech\Dissertation Stage 2 - Final\Final Character\Indian\P\0.jpg"
preds = model_predict(file_path, model)
img = image.load_img(file_path, target_size=(224, 224))
plt.imshow(img)
plt.show()
print("Predicted as: ", preds)
```

```python
def segment_hand(frame, threshold=25):
    global background
    
    diff = cv2.absdiff(background.astype("uint8"), frame)
    _ , thresholded = cv2.threshold(diff, threshold, 255,cv2.THRESH_BINARY)
    
     #Fetching contours in the frame (These contours can be of hand or any other object in foreground) …
    image, contours, hierarchy =cv2.findContours(thresholded.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    # If length of contours list = 0, means we didn't get any contours...
    if len(contours) == 0:
        return None
    else:
        # The largest external contour should be the hand 
        hand_segment_max_cont = max(contours, key=cv2.contourArea)
        
        # Returning the hand segment(max contour) and thethresholded image of hand...
        return (thresholded, hand_segment_max_cont)
```
